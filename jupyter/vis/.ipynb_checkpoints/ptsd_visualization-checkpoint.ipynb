```python
m = Sequential()
m.add(Dense(20,  activation='elu', input_shape=(df.shape[1],)))
m.add(Dense(2,    activation='linear', name="bottleneck"))
m.add(Dense(20  activation='elu'))
m.add(Dense(df.shape[1],  activation='sigmoid'))
m.compile(loss='mean_squared_error', optimizer = Adam())
history = m.fit(df[features], df[features], batch_size=128, epochs=5, verbose=1)

encoder = Model(m.input, m.get_layer('bottleneck').output)
Zenc = encoder.predict(df[features])
Renc = m.predict(df[features])
ax.scatter(Renc[:, 0], Renc[:, 1])
```
